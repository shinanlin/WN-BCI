```python
import sys

sys.path.append('../compare')

import numpy as np
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import ticker
from matplotlib.gridspec import GridSpec
from tqdm import tqdm
import pandas as pd
import os
```

```python
compare = []
add = '../results/compare'
subs = os.listdir(add)

for sub in subs:
    subFolder = add+os.sep+sub
    if os.path.exists(subFolder+os.sep+'NaB_info.csv'):
        f = pd.read_csv(subFolder+os.sep+'NaB_info.csv')
        compare.append(f)
    
compare = pd.concat(compare, ignore_index=True,axis=0)
compare.tail()
```

```python
sns.set_theme(style='ticks')

list = ['yangyuxing', 'jingyi','xuwenjia',
        'fanzixiao', 'wangchaoying', 'lijiayang', 
        'libowen', 'zhouyuqing', 'miaoyining', 
        'yaofeifan','chenyuzhen','shixinge',
        'suxiaohan','quyuxin','zhaoxinyu',
        'jiaqi', 'hanyue', 'weiyichen','liangfangshuo']

f,(ax1,ax2) =plt.subplots(1,2,figsize=(6,2.5))

plt.subplots_adjust(wspace=0.3)

sns.lineplot(data=compare[(compare.subject.isin(list))],
             x='f', y='ubSNR', hue='tag',ax=ax1)

sns.lineplot(data=compare[(compare.subject.isin(list))],
             x='f', y='ubrate', hue='tag',ax=ax2)


ax1.set_xlabel('SNR')
ax1.set_xlabel('Info(bit)')
```
