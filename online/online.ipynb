```python
import sys
sys.path.append('..')
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import pandas as pd
```

```python
cued = []
free = []
add = '../results/online'
subs = os.listdir(add)

for sub in subs:
    subFolder = add+os.sep+sub
    if os.path.exists(subFolder+os.sep+'cuedPerformance.csv'):

        f = pd.read_csv(subFolder+os.sep+'cuedPerformance.csv')
        cued.append(f)

        f = pd.read_csv(subFolder+os.sep+'freePerformance.csv')
        free.append(f)

cued = pd.concat(cued, ignore_index=True,axis=0)
free = pd.concat(free, ignore_index=True,axis=0)

cued.head()
```

## Describe Cued spelling

```python
cued.groupby(['subject','winLEN']).describe()[['ITR','accuracy']]
```
